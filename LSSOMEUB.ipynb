```python
%matplotlib inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from matplotlib.pylab import rcParams
from IPython.display import IFrame
rcParams['figure.figsize'] = 15, 10 # Make figures a little bigger
from tabulate import tabulate
```

```python
df = pd.read_csv('LS_SOM_EUB.csv') # Read file 
pd.to_datetime(df.Year, format='%Y') # Parse date as datetime
ts = df.set_index(df.Year) # set datetime as index
print(ts.head())
```

```python
fig, ax = plt.subplots()
plt.plot(ts.EUB, color='r', linestyle='-', label='Eurobarometer, Very satisfied')
plt.plot(ts.SOM, color='b', linestyle='-', label='SOM-institute, Very satisfied')
plt.ylim(0,1)
plt.legend(loc=2, borderaxespad=0.)
for k, v in ts.EUB.items():
    ax.annotate(int(round((v * 100))),xy=(k, v))
for k, v in ts.SOM.items():
    ax.annotate(int(round((v * 100))),xy=(k, v))
```
